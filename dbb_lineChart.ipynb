```
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from pylab import rcParams
rcParams['figure.figsize'] = 16, 8
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
results = pd.read_csv('/content/drive/My Drive/data/capgMyo/history-adaptedRNN-mixed.csv')
results
```

```
results['validation accuracy'] = results['validation accuracy'].apply(lambda x: x*100)
pivo = pd.pivot_table(results, index='Number of Trials (available for domain adaptation out of 5)', columns='Method', values='validation accuracy')
pivo
```

```
plt.rcParams.update({'font.size': 22})
plt.rc('xtick', labelsize=22)
plt.rc('ytick', labelsize=22)
plt.rc('legend', fontsize=22)
myFig = pivo.plot(linewidth=7.0)
plt.ylabel("Recognition Accuracy (%)", fontsize=22)
plt.xlabel("Number of Trials (available for domain adaptation out of 5)", fontsize=22)
ax = myFig.axes
ax.set_axis_bgcolor('white')
ax.grid(linewidth='0.2', color='grey')
start, end = ax.get_xlim()
ax.xaxis.set_ticks(range(1, 6, 1))
ax.yaxis.set_ticks(np.arange(0, 110, 10))
plt.savefig('/content/drive/My Drive/data/capgMyo/lineChart.pdf', format='pdf', dpi=1000, bbox_inches='tight')
```
